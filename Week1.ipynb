```python
df = pd.read_csv("data/소상공인시장진흥공단_상가업소정보_의료기관_201909.csv",low_memory=False)
df.shape
```

```python
df
```

```python
df.info()
```

```python
null_count = df.isnull().sum()
null_count
```

```python
df_null_count = null_count.reset_index()
df_null_count
```

```python
df_null_count.columns = ["컬럼명", "결측지수"]
df_null_count
```

```python
df_null_count_top = df_null_count.sort_values(by="결측지수", ascending=False).head(10)
df_null_count_top
```

```python
drop_columns = df_null_count_top["컬럼명"].tolist()
drop_columns
```

```python
df = df.drop(drop_columns, axis=1)
print(df.shape)
```

```python
df["시도명"]
```

```python
df_hospital = df[df["상권업종중분류명"] == "병원"].copy()
df_hospital
```

```python
hospital_count = df_hospital["시도명"].value_counts().reset_index()
hospital_count
```

```python
hospital_count.plot.bar("시도명")
```

```python
plt.figure(figsize = (16,12))
sns.scatterplot(data=df_hospital, x="경도", y="위도", hue="시도명")
```

<h1>강남에는 피부과가 다른 지역에 비해 많을까?</h1>

```python
df["상권업종소분류명"].unique()
```

```python
df_der = df[(df["상권업종소분류명"] == "피부과") & (df["시도명"] == "서울특별시")].copy()
df_der
```

```python
count_der = df_der["시군구명"].value_counts().reset_index()
count_der
```

```python
count_der.plot.bar("시군구명")
```

```python
gangnam_count = count_der[count_der["시군구명"] == "강남구"].copy()
other_count = count_der[count_der["시군구명"] != "강남구"]["count"].sum().copy()
```

```python
gangnam_count
```

```python
other_count
```

```python
new_df = pd.DataFrame({
    '시군구명': ['강남구', '그외 지역'],
    'count': [gangnam_count['count'].values[0], other_count]
})
```

```python
new_df
```

```python
new_df.plot.bar("시군구명")
```

```python
plt.figure(figsize = (9,8))
sns.scatterplot(data=df_der, x="경도", y="위도", hue="시군구명")
```
